```python
# Retrieve the content data into a dataframe
netflix_df = pd.read_csv("Resources/netflix_titles.csv")
netflix_df.head()
```

```python
#Clean data to contain only the required columns of the data related to the Analysis
data=netflix_df[['show_id','type','title','director','country','release_year','rating','duration','listed_in']]
data.head()
```

<img src="images/Data_analysis.jpg" height= "75" width="75" style="float:left;"/>

```python
#Count of movies and TV shows in total Netflix database. 
movies= data[data['type']=='Movie']['type'].count()
print(f"Total count of Movies in Netflix is {movies}.")
tv_show= data[data['type']=='TV Show']['type'].count()
print(f"Total count of TV shows in Netflix is {tv_show}.")
```

```python

typ= {'type':['Movies','Tv Shows'],
       'count':[movies,tv_show]}
typ=pd.DataFrame(typ)
plt.figure(figsize=(6,6))
plt.pie(typ['count'], labels=typ['type'],explode = (0, 0.09),colors=['red','gray'] ,autopct="%.2f%%", shadow=True, startangle=90)
plt.title("Movies vs TV Shows:")
plt.tight_layout()
```

<b><font color='black' size=3><tb><br>We see that there are 68.42% Movies and 31.58% TV Shows in Netflix. <br>  </font></b>   
<b><font color='black' size=5><tb> <br>-----------------------------------------------------------------<br> Movies Origination Data per Country: <br>-----------------------------------------------------------------<br> </font></b>
<b><font color='red' size=3><tb><br>Note: We will be doing the graphs for top 10 results only.  </font></b>   

```python
movies_df= data[data['type']=='Movie']
#(movies_df.head())

countriesm=movies_df['country'].value_counts()
plt.figure(figsize=(8,6))
plt.subplot(1,1,1)
plt.title("Count of movies per country")
plt.xlabel("Movies Count")
plt.ylabel("Country")
countriesm.head(10).plot(kind="barh", color='red')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('Count of movies per country.png')
plt.show()
```

<b><font color='black' size=3><tb>Interesting observation is that US has more than 1400 movies but India being in second place has less than 800 movies and UK has even less about 300 movies. <br></font></b>   
<b><font color='black' size=5><tb> <br>-----------------------------------------------------------------<br> TV Show Origination Data per Country: <br>-----------------------------------------------------------------<br> </font></b>
<b><font color='red' size=3><tb><br>Note: We will be doing the graphs for top 10 results only.  </font></b>   

```python
tv_show_df= data[data['type']=='TV Show']
countriess=tv_show_df['country'].value_counts()
plt.figure(figsize=(8,6))
plt.title("Count of Tv shows per country")
plt.xlabel("TV Show Count")
plt.ylabel("Country")

countriess.head(10).plot(kind="barh", color='red')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('Count of TV Shows per Country.png')
plt.show()
```

<b><font color='black' size=3><tb>
    <br>Observation from above graph is that US is still the top for TV Shows, But UK is in second place and India is not even in the top 5 for TV shows. <br> </font></b>

```python
Country_count = data.groupby(["country"]).count()["title"]
country_count_df = pd.DataFrame(Country_count)
sorted_df = country_count_df.sort_values(by=['title'],ascending=False)
sorted_df.reset_index(inplace=True)
sorted_df.head()
# Select top 5 countries with highest count of movies/shows
topcount = sorted_df.head(10)
#topcount.reset_index()

print(topcount.head(20))

```

    Movies / TV show releases per year:
<br>-------------------------------------------------------------------------------------------<br> </font></b>


```python
y_movies=movies_df['release_year'].value_counts()
y_tv_show=tv_show_df['release_year'].value_counts()

ym=pd.DataFrame(y_movies)

ym.reset_index(inplace=True)
ym.columns=['year','count']

yts=pd.DataFrame(y_tv_show)

yts.reset_index(inplace=True)
yts.columns=['year','count']
yearly=pd.merge(ym,yts, on='year',how="left")
yearly=yearly.rename(columns={'count_x':'movies_count','count_y':'tv_shows_count'})
yearly=yearly.dropna()
yearly['total']=yearly['movies_count']+ (yearly['tv_shows_count'])
```

```python
import seaborn as sns
sns.set(style="whitegrid")
f, ax = plt.subplots(figsize=(20, 15))
sns.set_color_codes("bright")
sns.barplot(x="year", y="total", data=yearly.head(20),
            label="Tv Shows", color="r")
sns.set_color_codes("bright")
sns.barplot(x="year", y="movies_count", data=yearly.head(20),
            label="Movies", color="k")


ax.legend(ncol=2, loc="upper left", frameon=True,fontsize='x-large')
plt.xticks(rotation=90, fontsize = 15)
sns.despine(left=True, bottom=True)
plt.savefig('Mv-TvS.png')
```
