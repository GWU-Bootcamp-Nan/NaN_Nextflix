```python
xls = pd.ExcelFile('Resources/statistic_id250934_number-of-netflix-paid-streaming-subscribers-worldwide-2011-2020.xlsx')

df = pd.read_excel(xls, 'Data')
df =df[['Unnamed: 1','Unnamed: 2']]
df=df.dropna()
subs=pd.DataFrame(df)
subs.columns=['Quartile','Number of Subscriber']
#subs
```

```python
his=pd.read_csv('Resources/NFLX_H.csv')
his=his.dropna()
hist=his['Close']
#hist
```

```python
y_values = hist
x_values = subs['Number of Subscriber']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.figure(figsize=(8,6))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(40,360),fontsize=15,color="red")
plt.title("Relation between Subsribers and Stock Value")
plt.ylabel("Stock Value")
plt.xlabel("Number of Subscribers(in millions)")
plt.tight_layout()
plt.show()
```

<b>Conclusion: Based on the number of subscribers growing per year and the stock value also rising constantly. We can assure that Netflix is a growing stock and is a very good place to invest your money. </b>
